```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load merged dataset
df = pd.read_csv("D:/courses/Internship/Task4_Disease_Risk_Prediction/OSBS_datasets/merged_final/combined.csv", parse_dates=["datetime"])
print("Shape:", df.shape)
print("\nBasic Info:")
df.info()
```

```python
df.head()
```

```python
# Sirf QF columns ko filter karein (jo naam mein "QF" rakhte hain)
qf_cols = [col for col in df.columns if "QF" in col]

# Function jo check karega ke row ke saare QF valid hain
def check_qf(row):
    for val in row[qf_cols]:
        if pd.notna(val):   # agar value NaN NAHI hai
            if val not in [0, 1]:  # aur 0 ya 1 ke ilawa kuch hai
                return False       # row hata do
    return True

# Apply filtering
df_clean = df[df.apply(check_qf, axis=1)]
```

```python
df_clean.shape
```

```python
df_clean.columns
```

```python
keep_cols = [
    "datetime","soilCO2concentrationMean","soilCO2concentrationMinimum","soilCO2concentrationMaximum","soilCO2concentrationVariance",
    "soilTempMean","soilTempMinimum","soilTempMaximum","soilTempVariance",
    "VSWCMean","VSWCMinimum","VSWCMaximum","VSWCVariance",
    "VSICMean","VSICMinimum","VSICMaximum","VSICVariance",
    "windSpeedMean","windSpeedMinimum","windSpeedMaximum","windSpeedVariance",
    "windDirMean","windDirVariance",
    "RHMean","RHMinimum","RHMaximum","RHVariance",
    "tempRHMean","tempRHMinimum","tempRHMaximum","tempRHVariance",
    "dewTempMean","dewTempMinimum","dewTempMaximum","dewTempVariance",
    "TFPrecipBulk"
]
# Filter dataset
df_reduced = df_clean[keep_cols]
```

```python
df_reduced.head()
```

```python
print("\nMissing Values (%):")
df_reduced.isnull().sum()
```

```python
# Make sure datetime column is in datetime format
df_reduced["datetime"] = pd.to_datetime(df_reduced["datetime"])

# Set datetime as index (needed for time-based interpolation)
df_reduced = df_reduced.set_index("datetime")

# Interpolate only numeric columns (up to 5 rows of missing values)
df_reduced = df_reduced.apply(
    lambda col: col.interpolate(method="time", limit=5) if col.dtype.kind in "fi" else col
)

# Drop remaining rows that still contain NaN values (large gaps)
df_reduced = df_reduced.dropna()
```
