```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("OSBS_datasets/Cleaned_data/NEON_30min_aggregated_data.csv",parse_dates=["datetime"])
```

```python
df.head()
```

```python
df.describe().T
```

```python
# Compute VPD
T = df['tempRHMean']  # Â°C
RH = df['RHMean']     # %
e_s = 0.6108 * np.exp((17.27 * T) / (T + 237.3))
e_a = (RH / 100.0) * e_s
df['VPD'] = e_s - e_a

# Create lag features (1h, 2h -> 2 and 4 steps because 30-min)
lag_vars = ['soilTempMean', 'VSWCMean', 'soilCO2concentrationMean', 'VPD']
for var in lag_vars:
    df[f'{var}_lag1h'] = df[var].shift(2)
    df[f'{var}_lag2h'] = df[var].shift(4)

df.fillna(df.mean(), inplace=True) 
```
